# Popularity Based Recommeder System

```python
ratings_with_book_name = books.merge(ratings, on = "ISBN")
```

```python
ratings_with_book_name
```

```python
ratings_with_book_name.shape
```

```python
ratings_with_book_name.sample(5)
```

```python
num_of_ratings = ratings_with_book_name.groupby("Book-Title")["Book-Rating"].count().reset_index()
num_of_ratings.rename(columns={"Book-Rating":"num of ratings"},inplace=True)
```

```python
num_of_ratings
```

```python
avg_book_rating = ratings_with_book_name.groupby("Book-Title")["Book-Rating"].mean().reset_index()
avg_book_rating.rename(columns={"Book-Rating":"Avg Rating"}, inplace=True)
```

```python
avg_book_rating
```

```python
popularity_df = num_of_ratings.merge(avg_book_rating,on="Book-Title")
```

```python
popularity_df 
```

```python
final_popular = popularity_df[(popularity_df["num of ratings"]>=250) & (popularity_df["Avg Rating"]>2)].sort_values("Avg Rating",ascending=False).head(50)
```

```python
final_popular
```

```python
popular_df = final_popular.merge(books,on="Book-Title")[["Book-Title","num of ratings","Avg Rating","Book-Author","Image-URL-M"]].drop_duplicates("Book-Title")
```
